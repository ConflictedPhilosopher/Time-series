```r
library(imputeTS)
library(astsa)
```

```r
data(jj)
plot(jj, main='Johnson & Johnson quarterly earnings per share-US', ylab='Earning/Share', xlab='Date', col='blue')
```

##### There is seasonality and trend in the time-series data. Thus the time series is not stationary.

```r
acf(jj, lag.max = NULL, type = "correlation", plot = TRUE, na.action = na.fail, demean = TRUE)
```

```r
# Removing the trend from the time series 
plot(diff(jj), main='Un-trended Johnson & Johnson quarterly earnings per share-US', ylab='Earning/Share', xlab='Date', col='blue')
acf(diff(jj), lag.max = NULL, type = "correlation", plot = TRUE, na.action = na.fail, demean = TRUE)
```
