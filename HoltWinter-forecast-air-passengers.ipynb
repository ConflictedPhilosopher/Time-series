```r
library(reshape)
library(forecast)
```

```r
par(mfrow=c(2,1))
plot.ts(AirPassengers, main='Number of Monthly Air Passengers (in thousands)')
plot.ts(log10(AirPassengers), main='log10:Number of Monthly Air Passengers (in thousands)')
```

```r
AirPassengers.HW = HoltWinters(log10(AirPassengers))
AirPassengers.HW$SSE
```

```r
# Fitted model parameters
AirPassengers.HW
```

```r
xhat <- AirPassengers.HW$fitted[, 1]
```

```r
require(graphics)

ts.plot(log10(AirPassengers), xhat,
        gpars=list(xlab="year", ylab="", lty=c(1:2)))
```

```r
AirPassengers.forecast <- forecast(AirPassengers.HW)
AirPassengers.forecast
```

```r
plot(AirPassengers.forecast, main='Forecast from HoltWinters')
```
