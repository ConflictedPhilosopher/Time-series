```r
library(tseries)
data(bev)
```

```r
plot(bev, ylab='Price', main='Beveridge Wheat Price Data')
bev.MA = filter(bev, rep(1/31, 31), sides=2)
lines(bev.MA, col='red')
```

#### To introduce some smothing a moving average filter is applied to the time series first

```r
par(mfrow=c(3,1))
Y = bev/bev.MA
```

```r
plot(Y, ylab='Scaled price', main='Transformed Beveridge Wheat Price Dat')
acf(na.omit(Y), main='ACF of Transformed Beveridge Data')
pacf(na.omit(Y), main='PACF of Transformed Beveridge Data')
```
